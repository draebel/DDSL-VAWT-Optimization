```python
import random
```

```python
def write_lines_to_file(lines, filename):
    f = open(filename,"w+")
    for line in lines:
         f.write(line + '\n')
    f.close() 
```

```python
for i in range(100):
    height = 0.2 + 4.8*random.random()
    score = min(height, 1/height)
    print("Aspect ratio: %s" % min(height, 1/height))
    lines = [
        "0 0", 
        "1 0", 
        "1 %s" % str(height),
        "0 %s" % str(height) 
    ]
    print(lines)
    write_lines_to_file(lines, "case=%s_score=%s.txt" % (height, score))
```
